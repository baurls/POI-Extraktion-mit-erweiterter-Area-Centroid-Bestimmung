## 2. Hilfsfunktionen

### Fläche berechnen

```python
def getAreaSize(points):
    A = 0
    for i in range(0, len(points)-1):
        A += (points[i][0]*points[i+1][1] - points[i+1][0]*points[i][1])
    A /= 2
    return A
```

### Polygone zufällig erstellen

```python
def generateRandomPoints(n, epsilon):
    points = []
    grad = (2*np.pi)/n
    currentAngle = 0
    radius = 5
    
    for i in range(n):
        x = radius * np.sin(currentAngle) 
        y = radius * np.cos(currentAngle) 
        points.append((x,y))
        currentAngle += grad
        
    new_points = distrubRandomly(points, epsilon)
    new_points.append(new_points[0])
    return new_points
```

```python
def spiltUpPoylgonRandomly(points):
    
    #presplit
    segments = []
    current_segment = []
    for point in points:
        current_segment.append(point)
        do_split = random();
        if do_split < 0.3 and len(current_segment) > 1:
            len(segments)
            segments.append(current_segment)
            current_segment = []
    segments[len(segments)-1].extend(current_segment)
    
    print(segments)
    
    
    #add same points
    first_run = True
    index = 0
    newSegments = []
    for segment in segments:
        if first_run:
            newSegments.append(segment)
            first_run = False
            index = index + 1
            continue
        conntecting_element = segments[index-1][len(segments[index-1])-1]
        index = index + 1
        newSegments.append([conntecting_element] + segment)
        
    return newSegments
```

```python
def distrubRandomly(points, epsilon):
    i = 0
    for (x,y) in points:
        x_rand = epsilon * random()
        y_rand = epsilon * random()
        points[i] = ((int)(10000*(x + x_rand)),(int)(10000*(y + y_rand)))
        i += 1
    return points
```
