## 2. Hilfsfunktionen

### Fläche berechnen

```python
def getTotalLength(points):
    l = 0.0
    for i in range(0, len(points)-1):
        lenSquared = (points[i][0] - points[i+1][0])**2 + (points[i][1] - points[i+1][1])**2
        l += lenSquared**0.5
    return l
```

### Geometrische Mitte berechnen

```python
def getLengthCentroid(points):
    #Points have format [(x0,y0), ... ,(x_n-1,y_n-1)]
    (x_s, y_s) =  (0,0)
    leftDistance = getTotalLength(points) / 2
    for i in range(0, len(points)-1):
        direction = getDirectionVector(points[i][0], points[i+1][0], points[i][1], points[i+1][1])
        length = getAbsolutLength(direction)
        
        if(length > leftDistance):
            alpha = 1 - (length - leftDistance)/ length
            newX = points[i][0] +  direction[0]*alpha
            newY = points[i][1] +  direction[1]*alpha
            return (newX, newY)
        leftDistance -= length

def getDirectionVector(xi, xj, yi, yj):
    #i = current
    #j = i+1
    return (xj - xi, yj - yi)

def getAbsolutLength((x,y)):
    return (x**2 + y**2)**0.5
```

### Centroid

```python
def getPointCentroid(points):
     #Points have format [(x0,y0), ... ,(x_n-1,y_n-1)]
    (x_s, y_s) =  (0.0,0.0)
        
    for (x,y) in points:
        (x_s, y_s) = (x_s + x, y_s + y)
            
    amount = (len(points))
    (x_s, y_s) =  (x_s/amount, y_s/amount)
    return (x_s, y_s)
```

### Polygone zufällig erstellen

```python
def generateRandomPathPoints(n, epsilon):
    points = []
    pathLength = 5
    currentX = 0.0
    
    #calc 'basic' path 
    for i in range(n):
        x = currentX
        y = 0.0
        points.append((x,y))
        currentX += pathLength
        
    #modify path randomly
    points = distrubRandomly(points, epsilon)
    return points
```

```python
def distrubRandomly(points, epsilon):
    i = 0
    print(epsilon)
    for (x,y) in points:
        x_rand = epsilon * random()
        y_rand = epsilon * random()
        points[i] = (x + x_rand, y + y_rand)
        i += 1
    return points
```

## 3. Spielwiese

```python
anzahlRunden = 4
anzahlPlots = 2
anzahlPunkte = 7
epsilon = 0.1

for disturb_level in range (1, anzahlRunden+1):
    distrubance = disturb_level * epsilon
    print('Störung: ' + str(distrubance) + 'eps')
    for round in range(anzahlPlots):
        points = generateRandomPathPoints(anzahlPunkte, distrubance)
       
        #An example for a different behaviour:
        #points = [(0,50), (1,-50), (2,50), (3,-50), (4,50), (5,-50), (6,50), (7,-50), (8,50), (9,0), (50,0)]

        x_koords = [x for (x,y) in points]
        y_koords = [y for (x,y) in points]
        
       
        (x_c, y_c) = getPointCentroid(points)
        print('Punkt aus Average-Centroid:' + str( (x_c, y_c)))
        (x_l, y_l) = getLengthCentroid(points)
        print('Punkt aus Längen-Centroid: ' + str( (x_l, y_l)))

        plt.plot(x_koords, y_koords, linewidth=2, solid_capstyle='round', zorder=2)
        plt.plot(x_l, y_l, 'xr')
        plt.plot(x_c, y_c, 'xg')
        plt.show()

        #Grün = Punkte zentrieren
        #Rot = Längenmitte
```
